```python
class Persona:
    def __init__(self,a,b): # ATTRIBUTI
        self.nome = a
        self.eta = b

    def scheda_attributi(self): # METODO
        return f"Scheda Persona:\n  Nome:{self.nome}\n  età:{self.eta}"

daniele = Persona('Daniele',36)

print(Persona.scheda_attributi(daniele))
```

```python
class Persona:
    def __init__(self,a,b): # ATTRIBUTI
        self.nome = a
        self.eta = b

    def scheda_attributi(self):
        return f"Scheda Persona:\n  Nome:{self.nome}\n  età:{self.eta}"

    def double(self,x): # METODO
        ''' moltiplica l'input x2 volte'''
        return x*2

    def triple(self,x): # METHOD
        ''' moltiplica l'input x3 volte'''
        return x*3

    def quadratoanni(self): # METHOD
        ''' RESTITUISCE IL QUADRATO DEGLI ANNI'''
        return self.eta**2

daniele = Persona('Daniele',36)
```

```python
daniele.double(23)
```

```python
marco.double(23)
```
